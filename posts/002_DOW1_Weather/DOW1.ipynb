```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
weather_df = pd.read_csv('philadelphia_weather_2005_to_2025.csv')
```

### I have always been one to enjoy my birthday, but that was when my birthday was spent back home in sunny Los Angeles. Ever since moving to Philadelphia for university, my birthday has fallen on rather gloomy days. When I thought about it the other day, I realized that it has rained on my birthday here in Philly ever since I got here. 

### I was curious to see how many times in the last 21 years it really did rain on February 27th!

![](bday.png)

## How many times has it rained on Feb 27th?

```python
feb27 = weather_df.query('day== "February 27"')
feb27
```

### As we can see here, in the last 21 years, it has rained 6 times on February 27th. Let's see if of those 6 times, three of them truly were during my college years:

```python
it_rained = feb27['rain']>0
feb27[it_rained]
```

### Aha! In my time at Penn (Aug 2022 - Present), it has in fact rained on my birthday every year! 

### Seems like the weather gods decided to take a year off in 2022, breaking up the streak from 2020-2025. 

### If only I had been born ten years earlier... would've seen blue skies and birthday smiles! Guess I am just cursed (just kidding)!

![](rain.png)

#### But how do we know this?

##### The steps taken via code are not all that different from what we would've done with a dataset by hand! The steps are actually rather simple:
- First, we filter for the February 27 dates in the 'day' column of the dataset
- Then, we access the 'rain' column to see if it rained or not. How do we know? If the listed number is > 0, it rained!
- Next, we count down the 'rain' column for the days with > 0.0 rain
- Finally, we concur: 6 recorded days of rain on February 27th in the last 21 years! 

## Now, how different would this be if I was born just a month later, say, on March 27th?

### Let's follow the same steps to find out!

```python
mar27 = weather_df.query('day== "March 27"')
mar27
```

### Now let's see how many of these March 27 days had recorded rain

```python
it_rained_march = mar27['rain']>0
mar27[it_rained_march]
```
