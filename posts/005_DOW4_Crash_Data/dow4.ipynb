
# Are Philadelphia’s Roads Getting Safer?  
### A look at 20 years of crash data

### Every day thousands of people move through Philadelphia by car, bike, bus, or on foot. Most trips are routine, but sometimes they end in crashes.

### Using PennDOT crash data from 2005–2024, I wanted to understand how traffic safety in Philadelphia has changed over time. At first glance the story seems simple: if crashes are declining, the roads must be getting safer. But the data ended up telling a more complicated story.


```python

import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,6)

df = pd.read_csv("PDOT_crash_data_2005_2024.csv", low_memory=False)

# Philadelphia County FIPS code
phl = df[df["COUNTY"] == 67].copy()

print("Total rows in dataset:", len(df))
print("Philadelphia crashes:", len(phl))
print("Years covered:", phl["CRASH_YEAR"].min(), "to", phl["CRASH_YEAR"].max())
```


## First question: are crashes actually going down?

If Philadelphia's roads are getting safer, the most obvious place to look is the number of crashes each year.

This felt like the simplest question to start with — but I had a feeling it wouldn't tell the whole story.


```python
crashes_per_year = phl.groupby("CRASH_YEAR").size()

crashes_per_year.plot()
plt.title("Total Crashes in Philadelphia (2005–2024)")
plt.xlabel("Year")
plt.ylabel("Number of Crashes")
plt.show()
```

At first glance this actually looks encouraging. The overall trend suggests that crashes have gradually declined over the past two decades.

But looking at this chart raised a new question for me.

If crashes are declining, does that automatically mean the roads are becoming safer?

Not necessarily. Fewer crashes could still involve serious injuries or deaths. So the next step was to look at crash severity.


## Are crashes becoming less deadly?

```python

fatalities = phl.groupby("CRASH_YEAR")["FATAL_COUNT"].sum()

fatalities.plot()
plt.title("Traffic Fatalities in Philadelphia (2005–2024)")
plt.xlabel("Year")
plt.ylabel("Fatalities")
plt.show()
```

This is where the story becomes more complicated.

While crashes generally decline, fatalities fluctuate much more dramatically. The two trends don’t move together as neatly as I expected.

So the story isn’t just about how many crashes happen. It also depends on when those crashes occur.

## When do crashes happen?
Cities have daily rhythms: morning commutes, school drop-offs, lunch traffic, evening rush hour, and late-night driving.

I wanted to see if crash patterns follow those rhythms too.

```python
hourly = phl.groupby("HOUR_OF_DAY").size()

hourly.plot(kind="bar")
plt.title("Crashes by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Crashes")
plt.show()

print("Most dangerous hour:", hourly.idxmax())
```

### Rush hour clearly matters. The busiest travel hours also tend to produce the most crashes.

### That observation made me think about who is most exposed in these environments. Drivers are protected by the vehicle around them, but pedestrians and cyclists are not.

### Who is most vulnerable?

```python
ped = phl.groupby("CRASH_YEAR")["PED_COUNT"].sum()
bike = phl.groupby("CRASH_YEAR")["BICYCLE_COUNT"].sum()

plt.plot(ped.index, ped.values, label="Pedestrians")
plt.plot(bike.index, bike.values, label="Cyclists")

plt.title("Pedestrian and Bicycle Involvement in Crashes")
plt.xlabel("Year")
plt.ylabel("People involved")
plt.legend()
plt.show()
```

#### These numbers are smaller than total crash counts, but they matter precisely because they represent people with far less protection.

#### At this point another question started to feel important: are crashes spread evenly across the city, or do they cluster in certain places?

## Are weekends actually more dangerous?

Weekends often feel riskier on the road — later nights, nightlife, and less predictable driving patterns.

But Philadelphia also has extremely heavy weekday commuting traffic. I was curious whether the data would reflect that.

So I looked at how crashes are distributed across the days of the week.

```python
day_labels = {
    1: "Sunday",
    2: "Monday",
    3: "Tuesday",
    4: "Wednesday",
    5: "Thursday",
    6: "Friday",
    7: "Saturday"
}

crashes_by_day = phl.groupby("DAY_OF_WEEK").size().rename(index=day_labels)

crashes_by_day.plot(kind="bar")

plt.title("Crashes by Day of Week in Philadelphia")
plt.xlabel("Day of Week")
plt.ylabel("Number of Crashes")

plt.xticks(rotation=45)

plt.show()
```

### This result pushes against some assumptions. While weekends might feel more chaotic on the road, weekday commuting patterns create a huge amount of traffic volume.

### In a dense city like Philadelphia, the daily rhythm of work, school, and commuting appears to play a major role in when crashes occur.


## Final Thoughts

#### Looking at two decades of crash data reveals a story that is both encouraging and complicated.

#### Total crashes in Philadelphia appear to have declined over time, suggesting improvements in vehicle safety, infrastructure, or driving behavior. But the data also shows that risk is not evenly distributed. Certain hours of the day, certain times of the year, and certain road users face greater danger.

#### What surprised me most is how the story changes depending on the question asked. Looking only at total crashes suggests progress, but examining severity, timing, and vulnerable road users reveals a much more complex picture of traffic safety in the city.
